```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
import folium

def color_pr(pr):
    if pr < 1000000: return 'green'
    elif 1000000 <= pr < 3000000: return 'orange'
    else: return 'red'
    
df_w = df[df.waterfront == 1]
map_w = folium.Map(location=[47.57,-122.3], zoom_start = 9.5)

for _  in range(df_w.shape[0]): 
    folium.CircleMarker(location=[df_w.lat.iloc[_], df_w.long.iloc[_]], radius = 2,\
                        color=color_pr(df_w.price.iloc[_])).add_to(map_w)
# сохранение карты в html
map_w.save('map2.html') 
display( map_w)
```
