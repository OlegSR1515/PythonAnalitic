<font size=2>Переведём кв.футы в привычные нам кв.метры для помещений и в сотки для участков :

```python
df_1 = df.copy()
```

```python
sq_apart = ['sqft_living', 'sqft_above', 'sqft_basement', 'sqft_living15']
sq_area = ['sqft_lot', 'sqft_lot15']

df_1[sq_apart] = df_1[sq_apart] / 10.7639
df_1[sq_area] = df_1[sq_area] / 10.7639 / 100

df_1.columns = df_1.columns.str.replace('ft', 'mt')
df_1.columns = df_1.columns.str.replace('sqmt_lot', 'sqsot_lot')
df_1.round().head(3)
```

```python
df_1.describe().round()
```

```python
df_51 = df[(df['waterfront'] != 0) & (df['bathrooms'] >= 3) & (df['sqft_basement'] != 0)]
print(F"5.1 Домов с видом на набережную, минимум с тремя ванными и с подвалом всего"
      F" {df_51.shape[0]} вариант.")
```

***

```python
df_52 = df[((df['view'] == 4) | (df['waterfront'] == 1)) &\
          ( (df['condition'] == 5) & (df['yr_built'] >= 1980) )]
print(F"5.2 Цена домов либо с очень красивым видом из окна, либо с видом на набережную,\n"
      F" в очень хорошем состоянии и годом постройки не меньше 1980 года\n находится в диапазоне от"
      F" ${df_52['price'].min():.0f} до ${df_52['price'].max():.0f}")
```

```python
df_53 = df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] < 150000)]
df_53[['sqft_basement', 'floors', 'price', 'condition']]   
```
