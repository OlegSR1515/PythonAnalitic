1.1 Создать новый признак price_per_sq_lot,<br>
который будет содержать среднюю стоимость за один кв. метр общей площади.<br>
1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах<br>
 между годом реновацией дома и годом постройки дома.<br>
Если реновации дома не было, то в новом признаке поставьте 0.<br>
1.3 Создайте признаки года продажи, месяца продажи.<br>

```python
import pandas as pd

df = pd.read_csv('kc_house_data.csv')
df.head()
```

***

<font size=3>_1.1._  <font size=2>Переведём кв.футы в привычные нам кв.метры.

```python
df_11 = df.copy()
df_11['price_per_sq_lot'] = df['price'] / df['sqft_lot'] / 10.7639
df_11.head()
```

***

<font size=3>_1.2._  <font size=2>Новый признак **delta_renovated**.

```python
df_12 = df.copy()
a = df['yr_built']
b = df['yr_renovated']
df_12['delta_renovated_1'] = [0 if b[_] == 0 else b[_] - a[_] for _ in range(df.shape[0])]
df_12[['yr_built','yr_renovated', 'delta_renovated_1']].head(4)
```

<font size=2>Или **так** :

```python
df_12['delta_renovated_2'] = df.apply(lambda x: 0 if x.yr_renovated == 0\
                                   else x.yr_renovated - x.yr_built, axis = 1)
df_12[['yr_built','yr_renovated', 'delta_renovated_1', 'delta_renovated_2']].head(3)
```

<font size=3>_1.3._  <font size=2>Признаки года продажи, месяца продажи.

```python
df_13 = df.copy()
mon = ['янв', 'фев', 'мар', 'апр', 'май', 'июн', 'июл', 'авг', 'сен', 'окт', 'ноя', 'дек']
```

```python
df_13['n_year_1'] = [df['date'][_][:4] for _ in range(df.shape[0])]
df_13['n_month_1'] = [int(df['date'][_][4:6]) for _ in range(df.shape[0])]

df_13['n_month_1'] = [mon[df_13['n_month_1'][_] - 1] for _ in range(df.shape[0])]
```

```python
df_13[['price', 'date','n_year_1', 'n_month_1']].head()
```

<font size=2>Или **так** :

```python
df_13['n_year_2'] = df.date.apply(lambda x: x[:4])
df_13['n_month'] = df.date.apply(lambda x: int(x[4:6]))

df_13['n_month_2'] = df_13.n_month.apply(lambda x: mon[x - 1])
sp = ['price', 'date','n_year_1', 'n_month_1','n_year_2', 'n_month_2']
df_13[sp].head()
```

***

<font size=3>_1.4._  <font size=2>Удалите признаки.

```python
df_14 = df.copy()
df_14.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df_14.head()
```

***

<font size=3>_2.1._  <font size=2> Метод **join**.

```python
clients_id = clients.set_index('house_id')
clients_id.head()
```

```python
hous_id = df.rename(columns = {'id':'house_id'}).set_index('house_id')
hous_id.head()
```

```python
d_j = clients_id.join(hous_id)
d_j.head()
```

***

<font size=3>_2.2._  <font size=2> Метод **merge**.

```python
d_m = clients_id.merge(df, left_on='house_id', right_on='id')
d_m.head()
```

***

<font size=2>Сравним обе таблицы. Для этого приведём их к одному формату:

```python
d_j_eq = d_j.copy()
d_j_eq.head(2) # Исходная таблица
```

```python
d_j_eq.reset_index(inplace= True) # Индекс установим как признак
d_j_eq = d_j_eq.sort_values(by=['client_id', 'price' ]) # Сортируем по двум столбцам(есть одинаковые значения)
d_j_eq = d_j_eq.reset_index(drop= True) # Индексируем заново
d_j_eq.head(4)
```

```python
c = d_j_eq.client_id
# Перемещаем колонку в начало
d_j_eq.drop(columns=['client_id'], inplace=True)
d_j_eq.insert(0, c.name, c)
d_j_eq.head(4) # Отформатированная таблица 
```

```python
d_m_eq = d_m.copy()# Исходная таблица 
d_m_eq.head(2)
```

```python
# Переименовываем столбец и сортируем, как выше
d_m_eq = d_m_eq.rename(columns = {'id':'house_id'}).sort_values(by=['client_id', 'price' ])
d_m_eq = d_m_eq.reset_index(drop= True) # Индексируем заново
d_m_eq.head(4) # Отформатированная таблица 
```

```python
d_m_eq.equals(d_j_eq)
```

<font size=3>_3.1._  <font size=2>Средняя стоимость домов в зависимости от количества спален.

```python
df.groupby('bedrooms').agg({'price':'mean'}).round()
```

***

<font size=3>_3.2._  <font size=2>Минимальная, средняя и максимальная стоимости домов
в зависимости от состояния дома.

```python
df.groupby('condition').agg({'price':['min', 'mean', 'max']}).round()
```

***

<font size=3>_3.3._ <font size=2>Количество домов в зависимости от вида на набережную и оценкой вида.

```python
df.pivot_table(index='waterfront', columns='view', values='date', aggfunc='count', fill_value=0)
```

<font size=2>Или **так** :

```python
df.groupby(['waterfront', 'view']).agg({'view' : 'count'})
```

***

<font size=3>_3.4._ <font size=2>Количество домов в зависимости от этажности и количества спален.

```python
pd.crosstab(index=df['floors'], columns=df['bedrooms'], margins=True, margins_name='Total')
```

***

<font size=3>_3.5._ <font size=2>Медианная стоимость домов в зависимости от состояния дома и оценки дома.

```python
df.pivot_table(index='condition', columns='grade', values='price', aggfunc='median', fill_value=0)
```

<font size=2>Или **так** :

```python
df.groupby(['condition', 'grade']).agg(
    {'price': 'median'}).unstack('grade').fillna(0)
```
