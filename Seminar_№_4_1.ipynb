```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<font size=3>_1.1._  <font size=2>Скачать данные :

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

<font size=3>_1.2._  <font size=2>Стоимость недвижимости :

```python
df.hist('price', bins= 80, color='m', alpha=0.5, edgecolor = 'k', grid= True, figsize=(11,4))
plt.title('Распределение стоимости недвижимости\n');
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

```python
df.boxplot('price', figsize=(11, 2), vert=False, whis=2)
plt.title('Cтоимость недвижимости');
```

<font size=3>_ВЫВОД :_ <br> <font size=2>1. Стоимость домов имеет очень большой разброс.<br> 
2. При этом **50 %** всех данных лежит в узком диапазоне **~ \\$300 000 - $600 000** (межквартильный диапазон).<br>
3. А больше **95,4 %** стоимости всех домов не дороже **~ \\$1 300 000** (**2-е сигмы**) .

<font size=3>_2.1._  <font size=2>Распределение домов от наличия вида на набережную :

```python
wt_count = df.waterfront.value_counts()
plt.figure(figsize=(11,6))
nam_w = ['Нет вида', 'Вид на набережную']

plt.title('Распределение домов от наличия вида на набережную :', c='b', fontsize=18)
plt.pie(wt_count, labels=nam_w, colors=['YellowGreen', 'Blue'], textprops={'fontsize':15},\
       explode=(0.4, 0), autopct='%1.1f%%', pctdistance=1.3, labeldistance=1.5) 

plt.show()
```

<div class="alert alert-block alert-success"><em>
    <b>Вывод:</b><br> Домов с вид на набережную крайне мало : всего <b>0,8 %</b> .
</em></div>

***

<font size=3>_2.2._  <font size=2>Распределение этажей домов :

```python
fl_count = df.floors.value_counts()
plt.figure(figsize=(18,5))
color = ['r', 'b', 'y', 'g', 'DeepPink', 'r']

plt.title('Распределение этажей домов :', c='g', fontsize=25)
plt.barh(fl_count.index, fl_count, color=color, height = 0.6)

for _ in range(len(fl_count)):
    i = fl_count.index[_]
    plt.text(fl_count[i] + 200,  i - 0.1, f'{fl_count[i]} ({fl_count[i] / 21613 * 100: 0.0f} %)',\
             color = 'Blue', fontsize = 20)

plt.show()
```

<div class="alert alert-block alert-success"><em>
<b>Вывод:</b><br> 96 % домов не выше 2-х этажей.
</em></div>

<font size=3>_2.3._  <font size=2>Распределение состояния домов :

```python
cn_count = df.condition.value_counts()
plt.figure(figsize=(11,6))
nam_c = cn_count.index
plt.pie(cn_count, labels=nam_c, textprops={'fontsize':15}, explode=(0, 0, 0.1, 0.3, 0.6),\
        wedgeprops={'edgecolor':'g', 'width':0.5}) #, hole=0.9
plt.title('Распределение состояния домов :', c='m', fontsize=18)
plt.show()
```

<font size=2>Категориальные признаки посмотрим на диаграммах размаха:

```python
df_vw = df[['price', 'view']].sort_values('view', ascending=False)
plt.figure(figsize=(18,5))
sns.boxplot(x=df_vw['price'], y=df_vw['view'].astype('str'), whis=3);
plt.xlabel('стоимость', fontsize=18)
plt.ylabel('вид из окон', fontsize=18)
plt.title('Зависимость стоимости от вида из окон', fontsize=22);
```

```python
df_wt = df[['price', 'waterfront']].sort_values('waterfront', ascending=False)
plt.figure(figsize=(18,5))
sns.boxplot(x=df_wt['price'], y=df_wt['waterfront'].astype('str'), whis=3);
plt.xlabel('стоимость', fontsize=18)
plt.ylabel('вид на набережную', fontsize=18)
plt.title('Зависимость стоимости от вида на набережную', fontsize=22);
```

```python
df_g = df[['price', 'grade']].sort_values('grade', ascending=False)
plt.figure(figsize=(18,5))
sns.boxplot(x=df_g['price'], y=df_g['grade'].astype('str'), whis=3);
plt.xlabel('стоимость', fontsize=18)
plt.ylabel('оценка дома', fontsize=18)
plt.title('Зависимость стоимости от оценка дома', fontsize=22);
```

```python
df_cn = df[['price', 'condition']].sort_values('condition', ascending=False)
plt.figure(figsize=(18,5))
sns.boxplot(x=df_cn['price'], y=df_cn['condition'].astype('str'), whis=5);
plt.xlabel('стоимость', fontsize=18)
plt.ylabel('состояние дома', fontsize=18)
plt.title('Зависимость стоимости от состояние дома', fontsize=22);
```

<div class="alert alert-block alert-success"><em>
<b>Вывод:</b><br> Из графиков видно, что виды из окон и на набережную и оценка дома сильно влияют на стоимость дома.<br>
А вот состояние дома хотя и влияет на стоимость, но очень слабо.
</em></div>

<font size=2>В заключение посмотрим на расположение домов:

```python
sns.jointplot(x=df.lat, y=df.long, kind='kde');
```

<font size=2>Из графика понятно, что дома расположены в одном месте, возможно, городе.<br>
Для разнообразия посмотрим, где расположены дома на берегу с разными ценами.

```python
import folium

def color_pr(pr):
    if pr < 1000000: return 'green'
    elif 1000000 <= pr < 3000000: return 'orange'
    else: return 'red'
    
df_w = df[df.waterfront == 1]
map_w = folium.Map(location=[47.57,-122.3], zoom_start = 9.5)

for _  in range(df_w.shape[0]): 
    folium.CircleMarker(location=[df_w.lat.iloc[_], df_w.long.iloc[_]], radius = 2,\
                        color=color_pr(df_w.price.iloc[_])).add_to(map_w)
# map_w    
```
